```python
batsman = pd.read_csv("c:/Users/hp/Downloads/sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
# ploting multiple plots
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
# lables title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title("Rohit Sharma vs Virat Kholi Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# Choose different colour
plt.plot(batsman['index'],batsman['V Kohli'],color="#1383eb")
plt.plot(batsman['index'],batsman['RG Sharma'],color='#f909a0')

plt.title("Rohit Sharma vs Virat Kholi Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="#1383eb", linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='#f909a0', linestyle='dashdot',linewidth=2)

plt.title("Rohit Sharma vs Virat Kholi Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="#1383eb", linestyle='solid',linewidth=3,marker='D')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#f909a0', linestyle='dashdot',linewidth=2,marker='.')

plt.title("Rohit Sharma vs Virat Kholi Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.grid()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color="#1383eb", linestyle='solid',linewidth=3,marker='D', label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#f909a0', linestyle='dashdot',linewidth=2,marker='.', label='Rohit')

plt.title("Rohit Sharma vs Virat Kholi Comparision")
plt.xlabel("Season")
plt.ylabel("Runs Scored")

plt.legend()
```
