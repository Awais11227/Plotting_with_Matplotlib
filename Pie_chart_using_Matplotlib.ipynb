# PieChart

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# simple data
data = [23,45,100,20,49]
subjects = ['English','Science','Maths','Statistics','Urdu']
plt.pie(data,labels=subjects)

plt.show()
```

```python
# dataset
df = pd.read_csv('c:/Users/hp/Downloads/gayle-175.csv')
df
```

```python
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',shadow=True)
plt.title('Batsman Runs Distribution')

plt.show()
```

```python
# percentage and colors
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',colors = ['#FF9999',  
          '#66B3FF', 
          '#99FF99',  
          "#FF001E",  
          '#FFB6C1',  
          '#8A2BE2'],  
          shadow=True   
)
plt.title('Batsman Runs Distribution')
plt.show()
```

```python
# explode shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1f%%',explode=[0.3,0,0,0,0,0.1],shadow=True)
plt.show()
```
